# **Ejemplo práctico**

### A continuación resolveremos un ejercicio práctico de Machine learning

```python
import tensorflow as tf
import numpy as np
```

## 2. Datos de entrenamiento


```python
epocas = 100

input = np.array(
#------------------------------------------------------------------
# Aquí abajo deben ir los datos que va a recibir el modelo  (USD) #|
#------------------------------------------------------------------
     [numero1, numero2, numero3, ...],
#------------------------------------------------------------------
     dtype=float)

out = np.array(
#----------------------------------------------------------------------------
# Aquí van los datos de muestra de lo que queremos recibir del modelo (COP) #|
#----------------------------------------------------------------------------
    [resultado_numero1, resultado_numero2, resultado_numero3, ...],
#----------------------------------------------------------------------------
    dtype=float)
```

## 3. Creación del modelo

nn.PNG

En estas líneas de código creamos nuestro modelo tal cual se observa en la imagen:

```python
oculta1 = tf.keras.layers.Dense(units=3, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([oculta1, oculta2, salida])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

### Entrenamiento del modelo

Este algoritmo se encarga de realizar el entrenamiento del modelo utilizando los datos de entrada que le entregamos:

```python
print("Comenzando entrenamiento...")
historial = modelo.fit(input, out, epochs= epocas, verbose=True)
print("Modelo entrenado!")
```

## 4. Análisis de errores

En estas líneas de código revisamos el comportamiento del modelo y que tan bien se ajusta a nuestros datos a lo largo de las diferentes iteraciones de entrenamiento:

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

## 5. Despliegue

Habiendo entrenado nuestro modelo, este está listo para realizar predicciones:

```python
print("Hagamos una predicción!")
predecir = 2
resultado = modelo.predict([predecir])
print(str(predecir)+" USD son = " + str(resultado[0][0])+"COP ")
```
